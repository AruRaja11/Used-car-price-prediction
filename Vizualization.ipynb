## Vizualization of car dekho

```python
## importing required packages

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
```

```python
## importing data

data = pd.read_csv("cleaned_details.csv")

data = data.iloc[:, 1:]

data.info()
```

```python
## reading raw data

raw_data = pd.read_csv('car_details.csv')

raw_data.head()
```

```python
## pairplot 

pairplot = data.iloc[:, :10]

sns.pairplot(pairplot)

plt.show()
```

```python
## distributions of features

plt.figure(figsize=(15, 12))  # Adjust figure size for better visibility
for idx, feature in enumerate(data.columns, 1):  # Start index from 1
    plt.subplot(6, 5, idx)  # Provide all three arguments (rows, cols, index)
    sns.histplot(data=data[feature], bins=20, kde=True)
    plt.title(feature)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
## Distribution of Price 

plt.figure(figsize=(15, 3))

sns.histplot(data=data['price'],bins=20, kde=True)
plt.show()
```

```python
## body_type distributions

top_data = raw_data['body_type'].value_counts()

plt.figure(figsize=(15, 3))
sns.barplot(x=top_data.values, y=top_data.index)
plt.title('Body Type vs Count')
plt.ylabel('Body Type')
plt.xlabel('Count')
plt.show()
```

```python
## correlation of columns

data_corr = data.corr()

plt.figure(figsize=(20, 10))
sns.heatmap(data=data_corr, annot=True, cmap='coolwarm')

plt.show()
```
