```python
# importing the required packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings

warnings.filterwarnings('ignore')
```

```python
#reading the dataframe
data = pd.read_csv('car_details.csv')
data1 = pd.read_csv('cleaned_details.csv')

data = data.iloc[:, 1:]
data1 = data1.iloc[:, 1:]
```

```python

data.dropna(subset='body_type', axis=0, inplace=True)

data.drop('price', axis=1, inplace=True)

data.reset_index(inplace=True)
data.drop('index',axis=1, inplace=True)

data['price'] = data1['price']
```

```python
# car body type per count of car
plt.figure(figsize=(13, 5))
sns.barplot(x=data['body_type'].values, y=data['body_type'].index)
plt.show()
```

```python
# histogram for the car brands
plt.figure(figsize=(30, 5))
sns.barplot(x=data['oem'].values, y=data['oem'].index)
plt.show()
```

```python
# plotting on fuel type, ownership, transmission and no of cylinders

fig, ax = plt.subplots(2, 2, figsize=(14, 5))
sns.barplot(x=data['fuel_type'].values, y=data['fuel_type'].index, label='Fuel Type', ax=ax[0, 0])
sns.barplot(x=data['owner'].values, y=data['owner'].index, label='Ownership', ax=ax[0, 1])
sns.barplot(x=data['transmission'].values, y=data['transmission'].index, label='Transmission', ax=ax[1, 0])
sns.barplot(x=data['No of Cylinder'].values, y=data['No of Cylinder'].index, label='No of Cylinder', ax=ax[1, 1])

plt.show()
```

```python
#vizualization of seats, gear box, kms driven, seating capacity

fig, ax = plt.subplots(2, 2, figsize=(13, 7))

sns.histplot(data=data1, x='Seats', label='Seating Capacity', kde=True, ax=ax[0, 0])
sns.kdeplot(data=data1, x='Kms Driven', label='Kms Driven', ax=ax[0, 1])
sns.histplot(data=data1, x='Gear Box', label='Gear Box', kde=True, ax=ax[1, 0])

plt.show()
```

```python
# plot between categorical columns and price

fig, ax = plt.subplots(2, 2, figsize=(13, 5))
sns.scatterplot(data=data1, x='Seats', y='price', ax=ax[0, 0])
sns.scatterplot(x=data['fuel_type'], y=data1['price'], ax=ax[0, 1])
sns.scatterplot(x=data['owner'], y=data1['price'],  ax=ax[1, 0])
sns.scatterplot(x=data['No of Cylinder'], y=data1['price'], ax=ax[1, 1])

plt.show()
```

```python
# average car price per car details 
fig, ax = plt.subplots(2,1, figsize=(40, 15))
sns.barplot(x=data['body_type'], y=data['price'], ax=ax[0])
sns.barplot(x=data['oem'], y=data['price'], ax=ax[1])
plt.show()
```

```python
#hist plot for categorical variables
fig, ax = plt.subplots(3, 3, figsize=(13, 6))
sns.histplot(data=data1, x='ownerNo', kde=True, label='Owner No',ax=ax[0, 0])
sns.histplot(data=data1, x='modelYear', kde=True, label='modelYear',ax=ax[0, 1])
sns.histplot(data=data1, x='price', kde=True, label='price',ax=ax[0, 2])
sns.histplot(data=data1, x='Seats', kde=True, label='Seats',ax=ax[1, 0])
sns.histplot(data=data1, x='Kms Driven', kde=True, label='Kms Driven',ax=ax[1, 1])
sns.histplot(data=data1, x='Displacement', kde=True, label='Displacement',ax=ax[1, 2])
sns.histplot(data=data1, x='No of Cylinder', kde=True, label='No of Cylinder',ax=ax[2, 0])
sns.histplot(data=data1, x='Gear Box', kde=True, label='Gear Box',ax=ax[2, 1])
sns.histplot(data=data1, x='Seats', kde=True, label='Seating Capacity',ax=ax[2, 2])

plt.show()
```

```python
#boxplot for numerical vairables

fig, ax = plt.subplots(2, 3, figsize=(15, 6))
sns.boxplot(data=data1, y='Length',ax=ax[0, 0])
sns.boxplot(data=data1, y='Width',ax=ax[0, 1])
sns.boxplot(data=data1, y='Height', ax=ax[0, 2])
sns.boxplot(data=data1, y='Wheel Base', ax=ax[1, 0])
sns.boxplot(data=data1, y='Kerb Weight', ax=ax[1, 1])


plt.show()
```

```python
#box plot for some other columns
fig, ax = plt.subplots(3, 3, figsize=(15, 6))
sns.boxplot(data=data1, y='ownerNo',ax=ax[0, 0])
sns.boxplot(data=data1, y='modelYear',ax=ax[0, 1])
sns.boxplot(data=data1, y='price', ax=ax[0, 2])
sns.boxplot(data=data1, y='Seats', ax=ax[1, 0])
sns.boxplot(data=data1, y='Kms Driven', ax=ax[1, 1])
sns.boxplot(data=data1, y='Displacement', ax=ax[1, 2])
sns.boxplot(data=data1, y='No of Cylinder', ax=ax[2, 0])
sns.boxplot(data=data1, y='Gear Box', ax=ax[2, 1])
sns.boxplot(data=data1, y='Seats', ax=ax[2, 2])

plt.show()
```

```python
# correlation 

col =['ownerNo', 'modelYear','Seats', 'Steering Type_electric', 'Kms Driven','Length', 'Width', 'Height', 'Top Speed', 'Wheel Base', 'Kerb Weight', 'Displacement', 'No of Cylinder','Gear Box','price']

correlation_matrix = data1[col].corr()

plt.figure(figsize=(13, 5))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('correlation heatmap')

plt.show()
```
