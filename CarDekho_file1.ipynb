## EXTRACTING DATA FROM THE RAW DATASETS ## 

```python
# importing packages

import pandas as pd
import numpy as np
import ast
```

```python
# reading the excel files

data1 = pd.read_excel("chennai_cars.xlsx")
data2 = pd.read_excel('bangalore_cars.xlsx')
data3 = pd.read_excel('delhi_cars.xlsx')
data4 = pd.read_excel('hyderabad_cars.xlsx')
data5 = pd.read_excel('jaipur_cars.xlsx')
data6 = pd.read_excel('kolkata_cars.xlsx')
```

```python
# converting the files to dataframe

chennai_df1 = pd.DataFrame(data1)
bangalore_df1 = pd.DataFrame(data2)
delhi_df1 = pd.DataFrame(data3)
hyderabad_df1 = pd.DataFrame(data4)
jaipur_df1 = pd.DataFrame(data5)
kolkata_df1 = pd.DataFrame(data6)
```

```python
# ['ignition_type', 'fuel_type', 'body_type', 'kilometers_driven', 'transmission', 'owner nos', 'owner', 'oem', 'model', 'modelyear', 'varientID', 'varient_name', 'price', 'priceActual', 'priceSaving', 'priceFixedText', 'trendingText']
col1 = eval(chennai_df1['new_car_detail'][0]).keys()
complete_cars = pd.DataFrame(columns=col1)
```

```python
# converting the codes in table to columns

def car_detail(location, loc, count):
    col1 = eval(chennai_df1['new_car_detail'][0]).keys()#evaluating the string 
    get_value = pd.DataFrame(columns=col1)

    for i in range(len(location['new_car_detail'])):
        temp = pd.DataFrame(eval(location['new_car_detail'][i]), columns=col1) #storing the result in dataframe
        temp['location'] = loc # adding location to the dataframe
        temp['index_number'] = count + i # index to remove the duplicated 
        
        get_value = pd.concat([get_value, temp], ignore_index=True) #concating the collected column value with the resultant
        
    get_value.drop_duplicates(subset=['index_number'], inplace=True)# droping the duplicates
    print(loc, ' Completed', " Total rows updated: ", i)
    return get_value
```

```python
# calling car_details function to collect data from column "new_car_detail"

location = [chennai_df1, bangalore_df1, delhi_df1, hyderabad_df1, jaipur_df1, kolkata_df1]
loc = ['chennai', 'bangalore', 'delhi', 'hydrabad', 'jaipur', 'kolkata']

count  = 0
for i in range(len(location)): #looping each states
    complete_cars = pd.concat([complete_cars, car_detail(location[i], loc[i], count)], ignore_index=True)

    count += 10

complete_cars.rename(columns={'it':'ignition_type', 'ft':'fuel_type', 'bt':'body_type', 'km':'kilometers_driven'}, inplace=True)
complete_cars
```

```python
# fetching datas from new_car_overview

def get_car_overview(data, loc):
    
    result = pd.DataFrame({}, columns=['result', 'location'])

    def get_data(dataframe):
        col = dataframe['result'][0].keys()
        temp = pd.DataFrame({}, columns=col)#creating the temp dataframe 
        for i in range(len(dataframe['result'])):# looping through string 
            val = pd.DataFrame(dataframe['result'][i].values())#fetching the values
            val = val.transpose().reset_index(drop=True)
            val.rename(columns={0:'Registration Year', 1:'Insurance Validity', 2:'Fuel Type', 3:'Seats', 4:'Kms Driven', 5:'RTO', 6:'Ownership', 7:'Engine Displacement', 8:'Transmission', 9:'Year of Manufacture'}, inplace=True)
            temp = pd.concat([temp, val], ignore_index=True)            
            
        return temp
            
    
    def extract(row): # extracting the values inside top key
        if row.get('top'):
            return {item['key']: item['value'] for item in row['top']} #returning the top as key value pair
        else:
            return None
    
    
    for j in range(len(data)):#looping through each row
        res = pd.DataFrame({'info': [str(data[j])]}) # result dataframe
    
        res['value_parsed'] = res['info'].apply(ast.literal_eval)
        
        res['result'] = res['value_parsed'].apply(extract)# calling extract function to extract data
        res['location']=loc
        result = pd.concat([result, res[['result', 'location']]]) #concating the dataframe
    result.reset_index(inplace=True)
    result.drop('index', axis=1, inplace=True)

    return get_data(result)
        
```

```python
## calling to get new_car_overview data

list = [data1['new_car_overview'], data2['new_car_overview'], data3['new_car_overview'], data4['new_car_overview'], data5['new_car_overview'], data6['new_car_overview']]

overview_df = pd.DataFrame()

for i in range(6):
    overview_df = pd.concat([overview_df, get_car_overview(list[i], loc[i])], ignore_index=True)#calling the result dataframe to combine with extracted value
    print(f"{i+1} state updated..")

overview_df
        
```

```python
# extracting data from new_car_spec
def get_top_spec(data, dataframe):
        
    def extract_top(row):# extracting the top value 
        if row.get('top'):
            top = {items['key']:items['value'] for items in row['top']} # returning the key values pairs of top
            return top
        return None 
        
    col = ['Engine', 'Max Power', 'Torque', 'Wheel Size', 'Seats'] #col names

    car_spec_final = pd.DataFrame({}, columns=col)
    
    spec_top = pd.DataFrame({}, columns=col)
    
    spec_data = pd.DataFrame({}, columns=col)
    
    for i in range(len(data)):#looping through the row
        car_spec = pd.DataFrame({
        'info':[data['new_car_specs'][i]]
        })
    
        car_spec['parsed'] = car_spec['info'].apply(ast.literal_eval)# evaluating the string 
        car_spec['top'] = car_spec['parsed'].apply(extract_top)# applying the extracting function to each row
        
        convert_top = pd.DataFrame(car_spec['top'][0].values())# extracting the top value
        convert_top = convert_top.transpose().reset_index(drop=True)
        convert_top.rename(columns={0:'Engine', 1:'Max Power', 2:'Torque', 3:'Wheel Size', 4:'Seats'}, inplace=True)
    
        spec_top = pd.concat([spec_top, convert_top], ignore_index=True)# concating the top value as a row to extract list value

    car_spec_final = pd.concat([car_spec_final, spec_top], ignore_index=True)

    return car_spec_final
```

```python
location = [chennai_df1, bangalore_df1, delhi_df1, hyderabad_df1, jaipur_df1, kolkata_df1]# state name


spec_tops = pd.DataFrame({}, columns=['Engine', 'Max Power', 'Torque', 'Wheel Size', 'Seats'])

for i in range(len(location)):# looping through states
    spec_tops= pd.concat([spec_tops, get_top_spec(location[i], spec_tops)], ignore_index=True)# concatng the extracted values to the dataframe
    print(f"dataframe {i+1} completed, {len(location[i])} updated")

spec_tops
```

```python
#columns to fetch the list value
def fetch_list(data):
    
    data_list = pd.DataFrame({}, columns=['value'])#creating the list value
    for i in range(len(data)):#looping the row
        testdf = pd.DataFrame({
            'info': [data['new_car_specs'][i]] # creating the tempdf to add each row
        })
        
        testdf['parsed'] = testdf['info'].apply(ast.literal_eval) #evaluating the string so that dict can be extracted
        
        all_list = []
        
        for sections in testdf['parsed']: #extracting the sections of top
            all_list.extend(sections['data']) #passing the data to list
    
        n = pd.DataFrame({
            'value': [",".join(map(str, all_list))]# combining the results from list to dataframe
        })
        data_list = pd.concat([data_list, n], ignore_index=True)#combining the results 
    return data_list
```
