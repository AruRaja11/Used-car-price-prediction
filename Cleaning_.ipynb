## CLEANING DATAFRAME ##

```python
# importing required packages
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
import re
```

```python
# reading the extracted dataframe 

car_details = pd.read_csv('car_details.csv', dtype={13: 'str'})

car_details.iloc[:, 30:40]
```

```python
# changing the datatype of the columns 

#1. ignition type 
car_details['ignition_type'] = car_details['ignition_type'].astype(int)

#2. kilometers_driven
car_details['kilometers_driven'] = car_details['kilometers_driven'].str.replace(",", "")
car_details['kilometers_driven'] = car_details['kilometers_driven'].astype(int)

#3. ownerNo
car_details['ownerNo'] = car_details['ownerNo'].astype(int)

#4. modelYear
car_details['modelYear'] = car_details['modelYear'].astype(int)

#5. centralVariantId
car_details['centralVariantId'] = car_details['centralVariantId'].astype(int)

#6. price
def get_price(row):
    try:
        try:
            pattern = r'\d+\.\d+'#creatign a pattern to extract price 
            result = re.search(pattern, str(row))# searching for pattern
            return float(result.group()) #returning the result 
        except AttributeError:
            pattern = r'\d+'# if the above pattern didn't found searching for this pattern
            result =re.search(pattern, str(row))
            return float(result.group())
        return 0
    except Exception as e:
        return 0
        
car_details['price'] = car_details['price'].apply(get_price)

#7.seats 
car_details['Seats'] = car_details['Seats'].apply(get_price)
car_details['Seats'] = car_details['Seats'].replace(0, None)

#8. Kmsdriven 
def getkms(row):
    try:
        pattern = r'\d+'
        result = re.search(pattern, str(row))
        return result.group()
    except Exception as e:
        return 0
car_details['Kms Driven'] = car_details['Kms Driven'].str.replace(",", "")
car_details['Kms Driven'] = car_details['Kms Driven'].apply(getkms)
car_details['Kms Driven'] = car_details['Kms Driven'].replace(0, None)

#9.Displacement
car_details['Displacement'] = car_details['Displacement'].fillna(0)
car_details['Displacement'] = car_details['Displacement'].replace(0, None)

#10.No of Cylinder
car_details['No of Cylinder'] = car_details['No of Cylinder'].fillna(0)
car_details['No of Cylinder'] = car_details['No of Cylinder'].replace(0, None)

#11.Gear Box
car_details['Gear Box'] = car_details['Gear Box'].fillna(0)
car_details['Gear Box'] = car_details['Gear Box'].str.replace(",", "")
car_details['Gear Box'] = car_details['Gear Box'].apply(getkms)
car_details['Gear Box'] = car_details['Gear Box'].replace(0, None)

#12.Front Tread 
car_details['Front Tread'] = car_details['Front Tread'].fillna(0)
car_details['Front Tread'] = car_details['Front Tread'].str.replace(",", "")
car_details['Front Tread'] = car_details['Front Tread'].apply(getkms)
car_details['Front Tread'] = car_details['Front Tread'].replace(0, None)

#13. Rear Tread
car_details['Rear Tread'] = car_details['Rear Tread'].fillna(0)
car_details['Rear Tread'] = car_details['Rear Tread'].str.replace(",", "")
car_details['Rear Tread'] = car_details['Rear Tread'].apply(getkms)
car_details['Rear Tread'] = car_details['Rear Tread'].replace(0, None)

#14.Top Speed
car_details['Top Speed'] = car_details['Top Speed'].fillna(0)
car_details['Top Speed'] = car_details['Top Speed'].apply(get_price)
car_details['Top Speed'] = car_details['Top Speed'].replace(0, None)

#15. Registration Year
car_details['Registration Year'] = car_details['Registration Year'].fillna(0)
car_details['Registration Year'] = car_details['Registration Year'].apply(getkms)

#16. length
car_details['Length'] = car_details['Length'].fillna(0)
car_details['Length'] = car_details['Length'].apply(get_price)
car_details['Length'] = car_details['Length'].replace(0, None)

#17. Width
car_details['Width'] = car_details['Width'].fillna(0)
car_details['Width'] = car_details['Width'].apply(get_price)
car_details['Width'] = car_details['Width'].replace(0, None)

#18. Height
car_details['Height'] = car_details['Height'].fillna(0)
car_details['Height'] = car_details['Height'].apply(get_price)
car_details['Height'] = car_details['Height'].replace(0, None)

#19. Wheel Base
car_details['Wheel Base'] = car_details['Wheel Base'].fillna(0)
car_details['Wheel Base'] = car_details['Wheel Base'].apply(get_price)
car_details['Wheel Base'] = car_details['Wheel Base'].replace(0, None)

#20. Kerb Weight
car_details['Kerb Weight'] = car_details['Kerb Weight'].fillna(0)
car_details['Kerb Weight'] = car_details['Kerb Weight'].apply(get_price)
car_details['Kerb Weight'] = car_details['Kerb Weight'].replace(0, None)

#20. No Door Numbers
car_details['No Door Numbers'] = car_details['No Door Numbers'].fillna(0)
car_details['No Door Numbers'] = car_details['No Door Numbers'].apply(get_price)
car_details['No Door Numbers'] = car_details['No Door Numbers'].replace(0, None)

print('data type changed')

```

```python
# cleaning datas in Steering Type
def steering_type(row):
    row = str(row)
    if (row.lower() == 'electric' or row.lower() == 'electrical' or row.lower() == "electronic" or row.lower() == 'epas'):
        return 'electric'
    elif (row.lower() == 'manual'):
        return 'manual'
    elif (row.lower() == 'power'):
        return 'power'
    else:
        return 'hydraulic'

car_details['Steering Type'] = car_details['Steering Type'].apply(steering_type)
```

```python
# cleaning insurance validity
def clean_insurance(row):
    if (row.lower() == 'petrol' or row.lower() == 'diesel' or row.lower() == 'electric'):
        return None
    elif row == 'Third Party insurance':
        return 'Third Party'
    elif (row == 'Not Available' or str(row) == '1' or str(row) == '2'):
        return None
    else: 
        return row

car_details['Insurance Validity'] = car_details['Insurance Validity'].apply(clean_insurance)
```

## ENCODING CATEGORICAL VARIABLES ##

```python
# Using one hot encoder for body_type 

ohe = OneHotEncoder(sparse_output=False)

label = LabelEncoder()
value = ohe.fit_transform(car_details[['body_type']])

features = ohe.get_feature_names_out(['body_type'])

body_type = pd.DataFrame(value, columns=features)

col = [car_details.columns] + [body_type.columns]

car_details = pd.concat([car_details, body_type], axis=1)

car_details.drop('body_type', axis=1, inplace=True)
```

```python
# Using lable encoding for location

label = LabelEncoder()

car_details['Location'] = label.fit_transform(car_details['location'])

car_details.drop('location', axis=1, inplace=True)
```

```python
# using label encoder for transmission

car_details['Transmission'] = label.fit_transform(car_details['transmission'])

car_details.drop('transmission', axis=1, inplace=True)
```

```python
# using one-hot encoder in Steering type

value = ohe.fit_transform(car_details[['Steering Type']])

features = ohe.get_feature_names_out(['Steering Type'])

steering_type = pd.DataFrame(value, columns=features)

col = [car_details.columns] + [steering_type.columns]

car_details = pd.concat([car_details, steering_type], axis=1)

car_details.drop('Steering Type', axis=1, inplace=True)
```

```python
# using one-hot encoder in fuel_type

value = ohe.fit_transform(car_details[['fuel_type']])

features = ohe.get_feature_names_out(['fuel_type'])

steering_type = pd.DataFrame(value, columns=features)

col = [car_details.columns] + [steering_type.columns]

car_details = pd.concat([car_details, steering_type], axis=1)

car_details.drop('fuel_type', axis=1, inplace=True)
```

```python
# using one-hot encoder for oem 

value = ohe.fit_transform(car_details[['oem']])

features = ohe.get_feature_names_out(['oem'])

manufacturer = pd.DataFrame(value, columns=features)

col = [car_details.columns] + [manufacturer.columns]

car_details = pd.concat([car_details, manufacturer], axis=1)

car_details.drop('oem', axis=1, inplace=True)
```

```python
# using one-hot encoder for insurance validity

value = ohe.fit_transform(car_details[['Insurance Validity']])

features = ohe.get_feature_names_out(['Insurance Validity'])

insurance = pd.DataFrame(value, columns=features)

col = [car_details.columns] + [insurance.columns]

car_details = pd.concat([car_details, insurance], axis=1)

car_details.drop('Insurance Validity', axis=1, inplace=True)
```
